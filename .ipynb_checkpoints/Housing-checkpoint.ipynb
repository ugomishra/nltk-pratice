```python
import pandas as pd
import hashlib as hs
import numpy as np

housing = pd.read_csv('./housing./housing.csv')
housingTrain = './housing./trainset.csv'
housingTest = './housing./testset.csv'

def checkHash(value, ratio):
    return hs.md5(np.int64(value)).digest()[-1] < 256*ratio

def dissectData(data, ratio, identifier):
    idList = 0*len(data)
    trainDf = []
    testDf = []
    print(identifier)
    if type(identifier) is list:
        for item in identifier:
            idList = data[item] + idList
    testIdsList = idList.apply(lambda id_: checkHash(id_, ratio))
    testDf = data.loc[testIdsList]
    trainDf = data.loc[~testIdsList]
    return testDf, trainDf
    

def splitData():    
    indexList = ['longitude','latitude', 'households']
    testDf, trainDf = dissectData(housing, 0.2, indexList)
    testDf.to_csv(housingTest)
    trainDf.to_csv(housingTrain)
    return trainDf.describe()
    
splitData()



```
